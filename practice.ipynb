#culimanate the user input and count

```python
value = float(input('Enter number for calculations purpose: '))
count = 0
total = 0

while value != 0:
  count = count +1
  total = total + value
  value = float(input('Enter another value or 0 to exit:'))

print('Total count:', count)
print('Total Sum: ', total)
```

#find the meximun value for a entered set of number

```python
value = float(input('Enter a number: '))

largest = value
while value != 0:
  if value > largest:
    largest = value
    print('New larget value:', largest)
  value = float(input('Enter another value or 0 to exit: '))

print('Largest value:', largest)
```
